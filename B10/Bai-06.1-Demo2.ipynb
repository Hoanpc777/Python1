```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
np.set_printoptions(precision=3, suppress=True)

print(pd.__version__)
```

# 1. Ghi/Đọc csv

```python
a = np.random.randint(0, 10, size=(10,4))
df1 = pd.DataFrame(data=a, index=range(1,11), columns=list('ABCD'), dtype='uint8')
df1
```

```python
df1.to_csv('data/df1.csv', index=True)
```

```python
df2 = pd.read_csv('data/df1.csv')
df2
```

```python
df2 = pd.read_csv('data/df1.csv', index_col=0)
df2
```

```python
df2 = pd.read_csv('data/df1.csv',  usecols=['A','C'])
df2
```

```python
df2 = pd.read_csv('data/df1.csv', index_col=0, usecols=[0,1,3])
df2
```

**Append một file csv**

```python
df3 = df1 * 2
df3.to_csv('data/df3.csv', mode='w', index=False)
```

```python
df1.to_csv('data/df3.csv', mode='a', index=False, header=False)
```

```python
# Cách khác: tự động xác định nếu file chưa có dữ liệu thì thêm header
# Lưu ý: Trên Windows có thể bị thêm dòng trống, khi đó cần set newline là rỗng
with open('data/df3.csv', mode='a', newline='') as f:
    df1.to_csv(f, index=False, header=f.tell()==0)
```

```python
pd.read_csv('data/df3.csv')
```

# 2. Ghi/Đọc Excel

```python
df1.to_excel('data/df1_excel.xlsx', sheet_name='DF1')
```

```python
# Mở rộng sheet mới
with pd.ExcelWriter('data/df1_excel.xlsx', engine='openpyxl', mode='a') as writer:
    df1.to_excel(writer, sheet_name='DF2')
```

```python
# Mở rộng sheet cũ? ==> Báo lỗi
with pd.ExcelWriter('data/df1_excel.xlsx', engine='openpyxl', mode='a') as writer:
    df1.to_excel(writer, sheet_name='DF1')
```

```python
# Có hướng dẫn dùng mode='A', nhưng không hoạt động!
with pd.ExcelWriter('data/df1_excel.xlsx', mode='A') as writer:
    df1.to_excel(writer, sheet_name='DF4')
```

```python
# Ghi vào nhiều sheets cùng lúc
with pd.ExcelWriter('data/df2_3_excel.xlsx') as writer:
    df1.to_excel(writer, sheet_name='DF2')
    df1.to_excel(writer, sheet_name='DF3')
```
