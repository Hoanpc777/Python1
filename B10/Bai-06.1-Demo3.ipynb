```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
np.set_printoptions(precision=3, suppress=True)

print(pd.__version__)
```

# 1. Biểu đồ điểm/đường

```python
a = np.array([3,1,3,4,5,2,6,5,8,3,9,1])
sa = pd.Series(a)

fig = plt.figure(figsize=(5,2))
sa.plot()
plt.show()
plt.close("all")
```

```python
fig = plt.figure(figsize=(5,2))
sa.plot(linestyle='' , marker='D', color='red')
plt.show()
plt.close("all")
```

```python
b = np.random.randint(4, 10, size=(10,3))
df_b = pd.DataFrame(b, index=range(1,11), columns=list('ABC'))

fig = plt.figure(figsize=(5,2))
ax = fig.add_subplot()
df_b.plot(linestyle='-' , marker='D')
plt.show()
plt.close("all")
```

```python
fig = plt.figure(figsize=(5,2))
ax = fig.add_subplot()
df_b.plot(ax=ax, linestyle='-' , marker='D')
plt.show()
plt.close("all")
```

```python
fig = plt.figure(figsize=(5,4))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
sa.plot(ax=ax1)
df_b.plot(ax=ax2, linestyle='-' , marker='D')
plt.tight_layout()
plt.show()
plt.close("all")
```

```python
df_b.plot(subplots=True, linestyle='-' , marker='D', sharex=True, figsize=(5,5))
plt.show()
plt.close("all")
```

# 2. Biểu đồ cột và histogram

```python
fig = plt.figure(figsize=(6,3))
ax = fig.add_subplot()
df_b.plot.bar(ax=ax, legend='reverse')
ax.set_xticklabels(range(1,11), rotation='horizontal')
ax.legend(loc='lower left')
plt.show()
plt.close("all")
```

```python
fig = plt.figure(figsize=(6,3))
ax = fig.add_subplot()
df_b.plot.bar(ax=ax, stacked=True, legend='reverse', rot=0)
plt.show()
plt.close("all")
```

```python
fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot()
df_b.plot.bar(ax=ax, stacked=True, legend='reverse', rot=0)
ax.bar_label(container = ax.containers[0], padding=-12)
ax.bar_label(container = ax.containers[1], padding=-12)
ax.bar_label(container = ax.containers[2], padding=-12)
ax.set_ylim([0,26])
#ax.set_xlim([0,11])
plt.legend(loc='upper left')
plt.show()
plt.close("all")
```

```python
fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot()
df_b.plot.bar(ax=ax, stacked=True, legend='reverse', rot=0)
ax.bar_label(container = ax.containers[0], label_type='center')
ax.bar_label(container = ax.containers[1], label_type='center')
ax.bar_label(container = ax.containers[2], label_type='center')
ax.set_ylim([0,26])
#ax.set_xlim([0,11])
plt.legend(loc='upper left')
plt.show()
plt.close("all")
```

```python
fig = plt.figure(figsize=(6,3))
ax = fig.add_subplot()
df_b.plot.barh(ax=ax, stacked=True, legend='reverse', rot=0)
plt.show()
plt.close("all")
```

```python
fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot()
df_b.plot.barh(ax=ax, stacked=True, legend='reverse', rot=0, 
               width=0.8, color=['r', 'g', 'b'], alpha=0.5)
ax.bar_label(container = ax.containers[0], label_type='center')
ax.bar_label(container = ax.containers[1], label_type='center')
ax.bar_label(container = ax.containers[2], label_type='center')
plt.show()
plt.close("all")
```

```python
df_b.plot.bar(subplots=True, sharex=False, sharey=True, rot=0, figsize=(6,6))
plt.show()
plt.close("all")
```

```python
df_b.plot.bar(subplots=True, layout=(1,3), sharex=False, sharey=True, rot=0, figsize=(6,3))
plt.show()
plt.close("all")
```

```python
c = np.random.normal(10, 3, size=(500,4))
df_c = pd.DataFrame(c, columns=list('ABCD'))
df_c.tail()
```

```python
fig = plt.figure(figsize=(6,3))
ax = fig.add_subplot()
df_c.plot.hist(ax=ax, stacked=True, bins=20, rot=0)
plt.show()
plt.close("all")
```

```python
df_c.plot.hist(subplots=True, rot=0, alpha=0.5)
plt.show()
plt.close("all")
```

```python
df_c.plot.hist(subplots=True, layout=(1,4), rot=0, alpha=0.5, figsize=(10,2), sharey=True)
plt.show()
plt.close("all")
```

```python
df_c.plot.hist(subplots=True, layout=(2,2), rot=0, alpha=0.5, figsize=(5,5), sharey=True)
plt.show()
plt.close("all")
```

```python
df_c.hist(alpha=0.5, figsize=(6,4))
plt.show()
plt.close("all")
```

# 3. Biểu đồ scatter

```python
ax = df_c.plot.scatter(x='A', y='B', figsize=(5,3), label='Group-1')
df_c.plot.scatter(x='A', y='C', ax=ax, color="yellow", label='Group-2')
plt.legend(loc='upper right')
plt.show()
plt.close('all')
```

```python
df_c.plot.hexbin(x='A', y='B', figsize=(5,3), gridsize=15)
plt.show()
plt.close('all')
```

# 4. Biểu đồ tròn

```python
df_c[0:1]
```

```python
df_c[0:4].plot.pie(y='A', figsize=(3,3))
plt.show()
```

```python
df_c[0:4].plot.pie(y='A', figsize=(3,3), labels=['CN1', 'CN2', 'CN3', 'CN4'], 
                   autopct='%.2f%%',  legend=False, title='Dữ liệu cột A', ylabel='')
plt.show()
```

```python
df_c[0:4].plot.pie(subplots=True, figsize=(10,8), legend=False, autopct='%.2f%%')
plt.tight_layout()
plt.show()
```

```python
# Vẽ quạt với Series các giá trị có tổng < 1.0
sc = pd.Series([0.12, 0.1, 0.03, 0.08])
sc.plot.pie(labels=['CN1', 'CN2', 'CN3', 'CN4'])
plt.show()
```

```python
# Vẽ quạt với Series các giá trị có tổng < 1.0
sc = pd.Series([0.12, 0.1, 0.03, 0.08])
sc.plot.pie(labels=['CN1', 'CN2', 'CN3', 'CN4'], ylabel='', radius=2.0,
            labeldistance=0.75, pctdistance= 1.0, autopct='%.2f%%',
            shadow = True, startangle=30)
plt.show()
```

## 2.5. Vẽ biểu đồ với table

```python
fig = plt.figure(figsize=(6,3))
ax = fig.add_subplot()
df_b.plot.bar(ax=ax, stacked=True, legend='reverse', table=True)
plt.xticks([])
plt.show()
plt.close("all")
```

```python
# Thử với biểu đồ tròn
df_c[0:4].plot.pie(y='A', figsize=(5,5), labels=['CN1', 'CN2', 'CN3', 'CN4'], 
                   autopct='%.2f%%',  legend=False, title='Dữ liệu cột A', ylabel='',
                   table=np.round(df_c[0:4]['A'],2))
plt.show()
plt.close("all")
```
