## Load bộ dữ liệu về 'tips' có sẵn của seaborn

```python
sns.get_dataset_names()
```

```python
tip_df = sns.load_dataset('tips')
tip_df.columns = ['Tổng đơn', 'Tip', 'Giới tính', 'Hút thuốc', 'Ngày', 'Thời điểm', 'Số lượng']
tip_df 
```

# 1. Relational plots (biểu đồ quan hệ)

**Scatterplot**

```python
# fig = plt.figure(figsize=(7,5))
sns.scatterplot(data=tip_df, x='Tổng đơn', y='Tip', hue='Giới tính')
```

```python
# hue_order = ['Female', 'Male']
# fig = plt.figure(figsize=(7,5))
sns.scatterplot(data=tip_df, x='Tổng đơn', y='Tip', 
                hue='Giới tính', hue_order=['Female', 'Male'])
```

**Lineplot**

```python
fig = plt.figure(figsize=(7,5))
sns.lineplot(data=tip_df, x='Tổng đơn', y='Tip', 
             hue='Giới tính', palette=['r', 'g'])
```

**Relational Plot (Figure-level)** với hàm relplot

```python
# kind = 'line'
fig = plt.figure(figsize=(7,5)) # Không có tác dụng
sns.relplot(data=tip_df, kind='line', 
            x='Tổng đơn', y='Tip', hue='Giới tính')
```

```python
# kind = 'scatter'
# Thay đổi kích thước bằng height và aspect
sns.relplot(data=tip_df, x='Tổng đơn', y='Tip', hue='Giới tính', 
            height=5, aspect=1.4)
```

**Regression plot**

```python
# sns.lmplot (lm = 'linear model', Figure-level)
sns.lmplot(data=tip_df, x='Tổng đơn', y='Tip', hue='Giới tính')
```

```python
# Axes level
fig = plt.figure(figsize=(7,5))
sns.regplot(data=tip_df, x='Tổng đơn', y='Tip', color='r')
```

# 2. Distributional plots (các biểu đồ về phân bố dữ liệu)

**Thử nghiệm với phân bố chuẩn**

```python
# Tạo dữ liệu với phân bố chuẩn
np.random.seed(1)
test_data = pd.DataFrame(np.random.normal(5,1,(500,2)), columns=['a', 'b'])
test_data['c'] = np.random.normal(0,1,500) > 0
test_data.head()
```

**displot**

```python
# Vẽ phân bố chuẩn với hàm displot và tinh chỉnh thông số biểu đồ
# fig = plt.figure(figsize=(1,1))
sns.displot(data=test_data, x='a', color='r', kde=True, bins=20, 
            height=5, aspect=5/3)
```

```python
# Vẽ biểu đồ kde (hoặc hist) với hàm displot
# kde: kernel density estimate ("ước lượng mật độ hạt nhân/lõi")
sns.displot(data=test_data, x='a', kind='kde', hue='c', palette=['r','g'], 
            height=3, aspect=5/3)
```

```python
# Hàm 'distplot' sẽ bị loại bỏ
fig = plt.figure(figsize=(5,3))
sns.distplot(test_data['a'], color='r', hist=True, kde=True, bins=20)
```

**kdeplot (kernel density estimate)**

```python
# Vẽ biểu đồ kde với hàm kdeplot (axes-level)
sns.kdeplot(data=test_data, x='a', hue='c', shade=True)
```

**histplot**

```python
# Vẽ biểu đồ histogram thông thường với hàm histplot (axes-level)
fig = plt.figure(figsize=(6,4))
sns.histplot(test_data, x='a', kde=True, bins=20, hue='c')
```

**Biểu đồ dạng cummulative (tổng lũy kế/cộng dồn)**

Hàm ecdfplot()

```python
sns.ecdfplot(data=test_data, x='b', hue='c', hue_order=[True,False])
```

**Biểu đồ dạng phân bố dày (rugplot)**

```python
sns.rugplot(data=tip_df, x='Tổng đơn', y='Tip', hue='Giới tính')
```

```python
# Kết hợp dụng 'rugplot' với các biểu đồ khác
# sns.scatterplot(data=tip_df, x='Tổng đơn', y='Tip', hue='Giới tính')
sns.lmplot(data=tip_df, x='Tổng đơn', y='Tip', hue='Giới tính')
# sns.relplot(data=tip_df, x='Tổng đơn', y='Tip', hue='Giới tính')
sns.rugplot(data=tip_df, x='Tổng đơn', y='Tip', hue='Giới tính')
```

# 3. Categorical plots (Biểu đồ có tính phân loại)

## Biểu đồ cột

```python
# Vẽ biểu đồ trung bình tổng đơn theo ngày trong tuần, và phân biệt khách hàng theo giới tính
# Một số estimator: np.mean, np.median, np.sum
bieudo = sns.barplot(data=tip_df, x='Ngày', y='Tổng đơn', hue='Giới tính', 
                     errcolor='r', capsize=0.2,
                     estimator=np.mean)
```

```python
# Nếu không cần error_bar thì set ci=None
bieudo = sns.barplot(data=tip_df, x='Ngày', y='Tổng đơn', hue='Giới tính', 
                     errcolor='r', ci=None,
                     estimator=np.sum)
```

```python
# Vẽ biểu đồ trung bình tổng đơn theo ngày trong tuần, 
# và phân biệt khách hàng theo giới tính
# Một số estimator: np.mean, np.median
sns.barplot(data=tip_df, x='Giới tính', y='Tổng đơn', hue='Ngày', 
            errcolor='k', ci=90,
            estimator=np.mean)
plt.legend(loc=[1.1, 0.5])
```

## Biểu đồ dạng box - violin - point

```python
sns.boxplot(data=tip_df, x='Ngày', y='Tổng đơn', hue='Giới tính')
```

```python
# 'enhanced box' cho các dữ liệu lớn
# Tham khảo: https://seaborn.pydata.org/generated/seaborn.boxenplot.html
sns.boxenplot(data=tip_df, x='Ngày', y='Tổng đơn')
```

```python
sns.violinplot(data=tip_df, x='Ngày', y='Tổng đơn', hue='Giới tính')
```

```python
sns.pointplot(data=tip_df, x='Ngày', y='Tổng đơn')
```

## Biểu đồ dạng cột - scatter

```python
# Figure-level
sns.catplot(data=tip_df, x='Ngày', y='Tổng đơn', 
            height=4, aspect=6/4, 
            hue='Giới tính',
            jitter=False)
```

```python
# Biểu đồ dạng string: mảnh/dải (axes-level)
sns.stripplot(data=tip_df, x='Ngày', y='Tổng đơn')
```

```python
# Biểu đồ dạng beeswarm: tổ ong (axes-level)
sns.swarmplot(data=tip_df, x='Ngày', y='Tổng đơn', hue='Giới tính')
```

# ** Vấn đề cần lưu ý **

Các hàm figure-level không hỗ trợ ax, nên không vẽ chung theo cách thông thường được

```python
fig, ax = plt.subplots()
sns.displot(test_data, x="a", ax=ax, height=2, aspect=2)
sns.displot(test_data, x="b", ax=ax, height=2, aspect=2)
```

**Chỉ có thể vẽ trên các axes một cách chủ động với các hàm axes-level**

```python
fig, ax = plt.subplots()
sns.histplot(test_data, x="a", ax=ax, color='r')
sns.histplot(test_data, x="b", ax=ax, color='g')
```
