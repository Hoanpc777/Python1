```python
import csv
from matplotlib import pyplot as plt
import numpy as np
```

1. Vẽ phương trình y = ax + b (cho phép gán giá trị khác nhau của a,b)

```python
x = np.arange(-10,10,0.1)
a = input('Hãy nhập hệ số a = ')
b = input('Hãy nhập hệ số b = ')
def draw(a,b,x):
    plt.plot(x,float(a) * x + float(b))
    plt.show()
draw(a,b,x)
```

2. Đọc và Vẽ biểu đồ dữ liệu chứng khoán

```python
file_path = "data-stock/excel_acb.csv"

row_list = []
with open(file_path, newline='') as csv_file:
    data_csv = csv.reader(csv_file, delimiter=',')
    header = next(data_csv)
    for row in data_csv:
        row_list.append(row)

ds_open  = [float(row[7])  for row in row_list]
ds_high  = [float(row[8])  for row in row_list]
ds_low   = [float(row[9])  for row in row_list]
ds_close = [float(row[10]) for row in row_list]
```

```python
n_data = len(ds_open)
x = range(n_data)

plt.figure(figsize = (10,6))

plt.plot(x, ds_open[::-1], color="red", label="open-value")
plt.plot(x, ds_close[::-1], color="green", label = "close-value")

plt.title("Giá cổ phiếu ACB",)
plt.xlabel("Thời Gian")
plt.ylabel("Giá CP")

plt.legend()

plt.show()
```

3.
a) Vẽ biểu đồ dữ liệu chứng khoán với nhiều mã khác nhau

```python
file_path = "data-stock/excel_acb.csv"
file_path1 = "data-stock/excel_fpt.csv"

def read_filePath(file_path):
    ds_open = []
    with open(file_path, newline='') as csv_file:
        data_csv = csv.reader(csv_file, delimiter=',')
        header = next(data_csv)
        for row in data_csv:
            ds_open.append(float(row[7]))
    return ds_open

acb_open = read_filePath(file_path)
fpt_open = read_filePath(file_path1)

plt.figure(figsize = (12,6))

plt.plot(range(len(acb_open)), acb_open[::-1], color="red", label="acb-open")
plt.plot(range(len(fpt_open)), fpt_open[::-1], color="green", label = "fpt-open")

plt.title("Biểu đồ giá cổ phiếu mở cửa của một số công ty")
plt.xlabel("Thời Gian")
plt.ylabel("Giá CP")

plt.legend()

plt.show()
```

3. b) Tương tự bài 1, nhưng cho phương trình y=ax^2 + bx + c (-5 <= x <=5)

```python
x = np.arange(-10,10,0.1)
a = input('Hãy nhập hệ số a = ')
b = input('Hãy nhập hệ số b = ')
c = input('Hãy nhập hệ số c = ')
def draw(a,b,c,x):
    plt.plot(x,float(a) * x * x + float(b) * x + float(c))
    plt.show()
draw(a,b,c,x)
```
