# Phần 1. Giới thiệu

```python
import matplotlib
import matplotlib.pyplot as plt

print(matplotlib.__version__)
# print(matplotlib)
# print(plt)
```

## 1.1) Ví dụ

```python
# Cách sử dụng đơn giản
x = [1,2,3]
y = [6,4,5]
plt.plot(x,y)
plt.show()
```

```python
# Vẽ một đường thẳng nối 2 điểm?
A = [2, 6]
B = [5, 3]
plt.plot([A[0], B[0]], [A[1], B[1]])
plt.show()
```

```python
# Thay đổi hình dạng marker, màu sắc đường
# Thêm title và các nhãn trên trục x, y

plt.plot([1,2,3], [6,4,5], color="red", marker="o")
plt.title("Demo Vẽ Hình với Matplot.Pyplot")
plt.xlabel("Giá trị của x")
plt.ylabel("Giá trị của y")
plt.xticks(x)
plt.yticks(y)
plt.show()
```

```python
# Subplots
plt.subplot(1,2,1)
plt.plot([1,2,3], [6,4,5], color="red", marker="o")

plt.subplot(1,2,2)
plt.plot([A[0], B[0]], [A[1], B[1]], color="green", marker="^")
plt.show()
```

## 1.3) Cấu trúc của biểu đồ trong Matplotlib

**Vẽ biểu đồ trên từng axes với hàm vẽ của mỗi axes**

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot([1,2,3], [6,4,5], color="red", marker="o")

ax2 = fig.add_subplot(2, 2, 2)
ax2.plot([1,2,3], [16,4,15], color="green", marker="o")

ax3 = fig.add_subplot(2, 2, 3)
ax3.plot([10,15,30], [6,4,5], color="blue", marker="o")
# ax4 = fig.add_subplot(2, 2, 4)
plt.show()
```

**Vẽ biểu đồ trên từng axes với hàm vẽ của plt**

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
plt.plot([1,2,3], [6,4,5], color="red", marker="o")

ax2 = fig.add_subplot(2, 2, 2)
plt.plot([1,2,3], [16,4,15], color="green", marker="o")

ax3 = fig.add_subplot(2, 2, 3)
plt.plot([10,15,30], [6,4,5], color="blue", marker="o")
# ax4 = fig.add_subplot(2, 2, 4)
plt.show()
```

**Cho phép share trục**

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
plt.plot([1,2,3], [6,4,5], color="red", marker="o")

ax2 = fig.add_subplot(2, 2, 2, sharey=ax1)
plt.plot([1,2,3], [16,4,15], color="green", marker="o")

ax3 = fig.add_subplot(2, 2, 3, sharex=ax1)
plt.plot([10,15,30], [6,4,5], color="blue", marker="o")
# ax4 = fig.add_subplot(2, 2, 4)
plt.show()
```

# Phần 2. Sử dụng hàm plot()

```python
# from matplotlib import pyplot as plt

plt.plot([1,2,3], [6,4,5], 
         color="red", alpha=0.5,
         marker="^", markersize=14, 
         linestyle="dashed", linewidth=2)
plt.title("Demo Hàm plot()")
plt.xlabel("Giá trị của x")
plt.ylabel("Giá trị của y")
plt.xticks(x)
plt.yticks(y)

plt.gca().invert_xaxis()
#ax.invert_xaxis()

plt.show()
```

```python
# Nếu không cần đường kẻ
from matplotlib import pyplot as plt
plt.subplot(1,2,1)
plt.plot([1,2,3], [6,4,5], color="red", marker="o", linestyle="dashed")

plt.subplot(1,2,2)
plt.plot([1,2,3], [3,5,0], color="green", marker="^", linestyle="")
plt.show()
```

```python
# Kết hợp chỉ định color/marker/linestyle
plt.plot([1, 2, 3], [1, 2, 3], 'go-')
plt.plot([4, 5, 6], [1, 2, 3], 'r^--')
plt.show()
```

```python
# Kết hợp chỉ định color/marker/linestyle
x1,y1 = [1, 2, 3], [1, 2, 3]
x2,y2 = [4, 5, 6], [1, 2, 3]
plt.plot(x1, y1, 'r^', x2, y2, 'g-')
plt.show()
```

```python
# Gán nhãn và sử dụng legend
plt.plot([1, 2, 3], [1, 2, 3], 'go-', label='Loại 1')
plt.plot([4, 5, 6], [1, 2, 3], 'r^-', label='Loại 2')
plt.legend()
plt.show()
```

```python
# Gán nhãn và sử dụng legend
plt.plot([1, 2, 3], [1, 2, 3], 'go-', label='Loại 1')
plt.plot([4, 5, 6], [1, 2, 3], 'r^-', label='Loại 2')
plt.legend(labels=["Type-A", "Type-B"])
plt.show()
```

```python
# Gán nhãn và sử dụng legend
plt.plot([1, 2, 3], [1, 2, 3], 'go-', label='Loại 1')
plt.plot([4, 5, 6], [1, 2, 3], 'r^-', label='Loại 2')
plt.legend(loc="lower right", ncol=2)
plt.show()
```

```python
# Sửa vị trí của legend theo tọa độ
plt.plot([1, 2, 3], [1, 2, 3], 'go-', label='Loại 1')
plt.plot([4, 5, 6], [1, 2, 3], 'r^-', label='Loại 2')
plt.legend(loc=[0.38, 0.7], ncol=2, frameon=False)
plt.show()
```

## 3.1) Figure size và tight_layout

```python
fig = plt.figure(figsize=(5,2), constrained_layout=True)
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
#plt.tight_layout()
plt.show()
```

```python
fig,axes = plt.subplots(3, 2, figsize=(7,5), constrained_layout=False)

for i in range(3):
    for j in range(2):
        ax = axes[i][j]
        ax.plot([1,2,3], [6,4,5])

plt.show()
```

## 3.2) Các thao tác trên axis

```python
# Bật/tắt axis với plt 
plt.plot([1,2,3], [6,4,5], 
         color="red", alpha=0.5,
         marker="^", markersize=14, 
         linestyle="dashed", linewidth=2)
plt.title("Demo Hàm plot()")
plt.xlabel("Giá trị của x")
plt.ylabel("Giá trị của y")

#plt.axis("off")
plt.xticks([])
#plt.yticks([])
plt.show()
```

**Bật tắt khi share trục**

```python
# Bật/tắt axis với ax
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
plt.plot([1,2,3], [6,4,5], color="red", marker="o")

ax2 = fig.add_subplot(2, 2, 2, sharey=ax1)
plt.plot([1,2,3], [16,4,15], color="green", marker="o")
#ax2.set_axis_off()
#ax2.set_yticks([])
#ax2.get_yaxis().set_visible(False)

ax3 = fig.add_subplot(2, 2, 3, sharex=ax1)
plt.plot([10,15,30], [6,4,5], color="blue", marker="o")

plt.show()
```

```python
# Bật/tắt axis với ax
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
plt.plot([1,2,3], [6,4,5], color="red", marker="o")

ax2 = fig.add_subplot(2, 2, 2, sharey=ax1)
plt.plot([1,2,3], [16,4,15], color="green", marker="o")
#ax2.set_yticks([])
ax2.get_yaxis().set_visible(False)

ax3 = fig.add_subplot(2, 2, 3, sharex=ax1)
plt.plot([10,15,30], [6,4,5], color="blue", marker="o")
ax3.get_xaxis().set_visible(False)
plt.show()
```

```python
# Thiết lập limits
fig = plt.figure()
ax1 = fig.add_subplot()

ax1.plot([1,2,3], [6,4,5], 
         color="red", alpha=0.5,
         marker="^", markersize=10, 
         linestyle="dashed", linewidth=2)
plt.title("Demo Hàm plot()")
plt.xlabel("Giá trị của x")
plt.ylabel("Giá trị của y")
plt.xlim([0,10])
plt.ylim([0,10])
# ax1.set_xticks([1,3,5,7,9])
# ax1.set_xticklabels(["Thứ 2","Thứ 3","Thứ 4","Thứ 5","Thứ 6"])

# Đảo ngược xaxis
# ax1.invert_xaxis()
# plt.gca().invert_xaxis()

plt.show()
```

## 3.3) Hiển thị lưới

```python
plt.plot([1,2,3], [6,4,5], 
         color="red", alpha=0.5,
         marker="^", markersize=14, 
         linestyle="dashed", linewidth=2)
plt.title("Demo Hàm plot()")
plt.xlabel("Giá trị của x")
plt.ylabel("Giá trị của y")
plt.xlim([0,10])
plt.ylim([0,10])
plt.grid(linestyle="--", linewidth=0.5, color='.25')
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
plt.plot([1,2,3], [6,4,5], color="red", marker="o")

ax2 = fig.add_subplot(2, 2, 2, sharey=ax1)
plt.plot([1,2,3], [16,4,15], color="green", marker="o")
ax2.grid(linestyle="--", linewidth=0.5, color='.25')

ax3 = fig.add_subplot(2, 2, 3, sharex=ax1)
plt.plot([10,15,30], [6,4,5], color="blue", marker="o")

plt.show()
```
