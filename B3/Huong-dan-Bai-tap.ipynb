```python
import matplotlib
import matplotlib.pyplot as plt
import csv
```

```python
filename = 'data-stock/excel_acb.csv'
```

```python
gia_close_list = []

with open(filename, 'r') as f:
    data_reader = csv.reader(f)
    h = next(data_reader)
    print(h)
    
    #row = next(data_reader)
    #print(float(row[10]))
    for row in data_reader:
        gia_close = float(row[10])
        #print(gia_close)
        gia_close_list.append(gia_close)
```

```python
len(gia_close_list)
```

```python
plt.plot(gia_close_list[0:100])
plt.show()
```
