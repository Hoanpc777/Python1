1. Vẽ phương trình y = ax + b (cho phép gán giá trị khác nhau của a,b)

```python
#New variables
x= list(range(1,5))
#Create new func
def pt(a,b):
    return [a * i + b for i in x]
#Choose a,b as requested
y = pt(2,1.5)
#Plot
import matplotlib.pyplot as plt
plt.plot(x, y, color="red", marker="o")
plt.title("Vẽ phương trình y = ax + b")
plt.xlabel("Giá trị của x")
plt.ylabel("Giá trị của y")
plt.xticks(x)
plt.yticks(y)
plt.show()   
```

2. Đọc và Vẽ biểu đồ dữ liệu chứng khoán

```python
#import module
import matplotlib.pyplot as plt
import csv, os, numpy
#New variable
gia_open,gia_close,gia_high,gia_low = [],[],[],[]
# List files
path = 'data-stock'
lst = os.listdir(path)
#
lst1=list(enumerate([[i,j]for j in range(3) for i in range(2)]))
figure, axis = plt.subplots(2,3,figsize=(22,8), constrained_layout=True)
for i in range(6):
    fname = lst[i]
    p = lst1[i][1]
    with open('data-stock/'+ fname,'r',encoding='utf-8') as f:
        data_reader = csv.reader(f)
        header = next(data_reader)
        for i in data_reader:
            gia_open.append(float(i[header.index('<Open>')]))
            gia_high.append(float(i[header.index('<High>')]))
            gia_low.append(float(i[header.index('<Low>')]))
            gia_close.append(float(i[header.index('<Close>')]))
        axis[p[0]][p[1]].plot(gia_open[0:len(gia_open)],label = 'Open')
        axis[p[0]][p[1]].plot(gia_high[0:len(gia_high)],label = 'High')
        axis[p[0]][p[1]].plot(gia_low[0:len(gia_low)],label = 'Low')
        axis[p[0]][p[1]].plot(gia_close[0:len(gia_close)],label = 'Close')        
        axis[p[0]][p[1]].set_title('Mã cổ phiếu '+ fname[fname.index('_')+1:fname.index('_')+4])
plt.legend()
plt.show()
```

3.2. Vẽ phương trình y=ax^2 + bx + c (-5 <= x <=5)

```python
#New variables
x= list(range(-5,6))
#Create new func
def pt(a,b,c):
    for i in x:
        return [a *i**2 + b*i + c for i in x]
#Choose a,b,c as requested
y = pt(2,1.5,3)
#Plot
import matplotlib.pyplot as plt
plt.plot(x, y, color="red", marker="o")
plt.title("Vẽ phương trình y=ax^2 + bx + c")
plt.xlabel("Giá trị của x")
plt.ylabel("Giá trị của y")
plt.xticks(x)
plt.yticks(y)
plt.show()   
```
