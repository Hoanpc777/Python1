1. Vẽ phương trình y = ax + b (cho phép gán giá trị khác nhau của a,b)

```python
#New variables
x= list(range(1,5))
#Create new func
def pt(a,b):
    return [a * i + b for i in x]
#Choose a,b as requested
y = pt(2,1.5)
#Plot
import matplotlib.pyplot as plt
plt.plot(x, y, color="red", marker="o")
plt.title("Vẽ phương trình với Matplotlib")
plt.xlabel("Giá trị của x")
plt.ylabel("Giá trị của y")
plt.xticks(x)
plt.yticks(y)
plt.show()

    
```

2. Đọc và Vẽ biểu đồ dữ liệu chứng khoán

```python
#import module
import matplotlib.pyplot as plt
import csv, os
# List files
path = 'data-stock'
lst = os.listdir(path)
# lst_fname = ['{0}'.format(','.join(str(i))) for i in lst]
#
for i in lst:
    with open('data-stock/'+i,'r',encoding='utf-8') as f:
        data_reader = csv.reader(f)
        header = next(data_reader)
        gia_open = [float(i[header.index('<Open>')]) for i in data_reader]
    fig = plt.figure()
    plt.plot(gia_open[0:100])
    plt.title('Mã cổ phiếu '+ i[i.index('_')+1:i.index('_')+4])
    plt.show()
        
        
```
